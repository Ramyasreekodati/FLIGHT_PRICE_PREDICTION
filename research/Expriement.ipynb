```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot(df, col, primary_color='#FF005A', secondary_color='#ff6347',third_color="#FFA500"):
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 12))
    
    sns.kdeplot(df[col], ax=ax1, fill=True, color=primary_color)
    ax1.set_title(f'Distribution of {col}', color=primary_color)
    ax1.tick_params(colors=primary_color)
    
    sns.boxplot(x=df[col], ax=ax2, color=third_color)
    ax2.set_title(f'Boxplot of {col}', color=third_color)
    ax2.tick_params(colors=third_color)
    
    sns.histplot(df[col], ax=ax3, kde=False, color=secondary_color)
    ax3.set_title(f'Histogram of {col}', color=secondary_color)
    ax3.tick_params(colors=secondary_color)
    
    # Set the overall plot theme
    plt.tight_layout()
    plt.show()

# Call the function with the 'Charcoal and Mint' theme
plot(flight, "Price", primary_color='#FF005A', secondary_color='#ff6347',third_color="#FFA500")


```

```python
x=flight.drop(columns="Price",axis=1)
y=flight["Price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
ml_model=RandomForestRegressor(max_depth=30, max_features='sqrt', min_samples_split=10,n_estimators=400)
```

```python
ml_model.fit(X_train,Y_train)
```

```python
Y_pred=ml_model.predict(X_test)
Y_pred
```

```python
from sklearn import metrics
```
